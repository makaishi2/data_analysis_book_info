### 演習問題
このビデオ店では、販売促進施策として、過去のレンタル金額の合計が一番大きな顧客を
プラチナユーザーとし、そのユーザーが今まで一番大きなレンタル金額を支払っているカテゴリの
レンタル料を半額にすることとしました。  
この方針に基づき、次の２つの問いに答えて下さい。  
  
 (1) 過去の合計レンタル金額が最も高いユーザーを調べて下さい。  
 (2) (1)で調べたユーザーに対して、過去のレンタル金額が最も大きいカテゴリ名を調べて下さい。

#### (1) 過去の合計レンタル金額が最も高いユーザーを調べる

```python
# 貸出情報から項目「顧客ID」と「レンタル代」のみを抽出し、結果をx1に代入する
# 参照　2.4.4項　コード2-4-15　
x1 = df1[['顧客ID', 'レンタル代']]
display(x1.head(2))
```

```python
#  x1に対してgroupbyメソッドを用いて、レンタル代の合計金額を顧客IDごとに算出
# 参照 3.4.5項　コード3-4-11
x2 = x1.groupby('顧客ID').sum()
display(x2.head(2))
```

```python
# x2に対してreset_indexメソッドと用いて、顧客　IDを項目名に戻す
# 参照 3.6.4項　コード3-6-7
x3 = x2.reset_index()
display(x3.head(2))
```

```python
# x3の結果を「レンタル代」で逆順にソートして、
# レンタル代の合計が一番大きな顧客を算出
# 参照 3.6.3項コード3-6-4
x4 = x3.sort_values('レンタル代', ascending=False)
display(x4.head(2))
```

#### (2) (1)で調べたユーザーに対して、過去のレンタル金額が最も大きいカテゴリ名を調べる 

```python
# 貸出情報から項目「顧客ID」と「レンタル代」「カテゴリ名」のみを抽出し、
# 結果をx5に代入する
# 参照　2.4.4項　コード2-4-15　
x5 = df1[['顧客ID', 'レンタル代', 'カテゴリ名']]
display(x5.head(2))
```

```python
# x5から顧客IDが(1)で調べたものを抽出
# 参照 3.6.5項　コード3-6-9
x6 = x5.query('顧客ID == 148')
display(x6.head(2))
```

```python
# groupbyメソッドを用いてx6の結果から
# カテゴリ別の合計レンタル代を算出
# 参照 3.4.5項　コード3-4-11
x7 = x6.groupby('カテゴリ名').sum()['レンタル代']

# reset_indexメソッドを用いて、カテゴリ名を列に戻す
# 参照 3.6.4項　コード3-6-7
x8 = x7.reset_index()
display(x8.head(2))
```

```python
# sort_indexメソッドで、合計レンタル代の最も大きな
# カテゴリを調べる
# 参照 3.6.3項コード3-6-4
x9 = x8.sort_values('レンタル代', ascending=False)
display(x9.head(2))
```
